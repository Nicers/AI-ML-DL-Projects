```python
import os
import pandas as pd

# Load your Excel file
labels_df = pd.read_excel(
    r"C:\Users\USTB-SE\Documents\Python Research\SymDerm_v2_DB\SymDerm_v2_labels_with_all_malignancy_updated6.xlsx"
)

# Paths to your folders
base_path = r"C:\Users\USTB-SE\Documents\Python Research\SymDerm_v2_DB\rawDataset\dat\Lesion segmentation results\Ground truth"
datasets = ['DermIs', 'DermQuest']

# Dictionary to store normalized filename → label
label_dict = {}

# Fill dictionary from folder structure
for dataset in datasets:
    for label_folder, label_value, label_text in [('Melanoma', 1, 'Malignant'), ('Not Melanoma', 0, 'Benign')]:
        folder_path = os.path.join(base_path, dataset, label_folder)
        if not os.path.exists(folder_path):
            continue
        for filename in os.listdir(folder_path):
            normalized_name = filename.lower().replace('.jpg', '').replace('.jpeg', '')
            label_dict[normalized_name] = {
                'Malignancy_Labels': label_value,
                'Malignancy_Text': label_text
            }

# Update DataFrame rows
def update_row(row):
    file_name = row['File']
    normalized_file = file_name.lower().replace('_orig', '').replace('.jpg', '').replace('.jpeg', '')

    match = label_dict.get(normalized_file)
    if pd.isna(row['Malignancy_Labels']) and match:
        row['Malignancy_Labels'] = match['Malignancy_Labels']
        row['Malignancy_Text'] = match['Malignancy_Text']
    return row

# Apply updates
labels_df = labels_df.apply(update_row, axis=1)

# Save updated file
labels_df.to_excel(r"C:\Users\USTB-SE\Documents\Python Research\SymDerm_v2_DB\SymDerm_v2_labels_with_all_malignancy_updated.xlsx", index=False)
```

```python
def remove_hair(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (17, 17))
    blackhat = cv2.morphologyEx(gray, cv2.MORPH_BLACKHAT, kernel)
    _, mask = cv2.threshold(blackhat, 10, 255, cv2.THRESH_BINARY)
    inpainted = cv2.inpaint(image, mask, inpaintRadius=1, flags=cv2.INPAINT_TELEA)
    return inpainted

hair_removed_images = [remove_hair(img) for img in resized_images]
print("Hair removal completed.")
```

```python
import numpy as np
import cv2
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

# Load processed images and masks
augmented_images = np.load(r'C:\Users\USTB-SE\Documents\Python Research\Processed_Data\Result-2\augmented_modified_cropped_images.npy', allow_pickle=True)
masks_array = np.load(r'C:\Users\USTB-SE\Documents\Python Research\Processed_Data\Result-2\augmented_masks_array2.npy', allow_pickle=True)

# Normalize images (Scale pixel values to [0,1])
augmented_images = augmented_images.astype("float32") / 255.0

# Ensure masks have a single channel
if len(masks_array.shape) == 3:  # If masks are (N, H, W)
    masks_array = masks_array[..., np.newaxis]  # Reshape to (N, H, W, 1)

# Split data into training and validation sets (80% Train, 20% Validation)
X_train, X_val, y_train, y_val = train_test_split(augmented_images, masks_array, test_size=0.2, random_state=42)

# Target size
IMG_SIZE = (224, 224)

# Resize training images
X_train_resized = np.array([cv2.resize(img, IMG_SIZE) for img in X_train])
X_val_resized = np.array([cv2.resize(img, IMG_SIZE) for img in X_val])

# Resize masks
y_train_resized = np.array([cv2.resize(mask, IMG_SIZE, interpolation=cv2.INTER_NEAREST) for mask in y_train])
y_val_resized = np.array([cv2.resize(mask, IMG_SIZE, interpolation=cv2.INTER_NEAREST) for mask in y_val])

# Number of classes (Adjust based on segmentation labels)
num_classes = 3

# Ensure mask values are within the correct range (0, 1, 2)
y_train_resized = (y_train_resized / 255 * (num_classes - 1)).astype(np.uint8)
y_val_resized = (y_val_resized / 255 * (num_classes - 1)).astype(np.uint8)

# One-hot encode the masks
y_train_final = to_categorical(y_train_resized, num_classes=num_classes)
y_val_final = to_categorical(y_val_resized, num_classes=num_classes)

# Print final shapes
print(f"X_train shape: {X_train_resized.shape}, y_train shape: {y_train_final.shape}")
print(f"X_val shape: {X_val_resized.shape}, y_val shape: {y_val_final.shape}")

# Use these variables for training
X_train, X_val, y_train, y_val = X_train_resized, X_val_resized, y_train_final, y_val_final
```

```python
from tensorflow.keras.layers import BatchNormalization, LeakyReLU, Dropout
from tensorflow.keras.activations import swish
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Conv2DTranspose, concatenate, Input
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
# from tensorflow.keras.metrics import MeanIoU

def unet_model(input_shape=(224, 224, 3)):
    inputs = Input(input_shape)

    # Encoder
    c1 = Conv2D(64, (3, 3), padding='same', kernel_initializer='he_normal')(inputs)
    c1 = BatchNormalization()(c1)
    c1 = LeakyReLU(negative_slope=0.1)(c1)
    c1 = Conv2D(64, (3, 3), padding='same', kernel_initializer='he_normal')(c1)
    c1 = BatchNormalization()(c1)
    c1 = LeakyReLU(negative_slope=0.1)(c1)
    p1 = MaxPooling2D((2, 2))(c1)

    c2 = Conv2D(128, (3, 3), padding='same', kernel_initializer='he_normal')(p1)
    c2 = BatchNormalization()(c2)
    c2 = LeakyReLU(negative_slope=0.1)(c2)
    c2 = Conv2D(128, (3, 3), padding='same', kernel_initializer='he_normal')(c2)
    c2 = BatchNormalization()(c2)
    c2 = LeakyReLU(negative_slope=0.1)(c2)
    p2 = MaxPooling2D((2, 2))(c2)

    # Bottleneck
    c3 = Conv2D(256, (3, 3), padding='same', kernel_initializer='he_normal')(p2)
    c3 = BatchNormalization()(c3)
    c3 = LeakyReLU(negative_slope=0.1)(c3)
    c3 = Dropout(0.5)(c3)  # Increased Dropout for better regularization
    c3 = Conv2D(256, (3, 3), padding='same', kernel_initializer='he_normal')(c3)
    c3 = BatchNormalization()(c3)
    c3 = LeakyReLU(negative_slope=0.1)(c3)

    # Decoder
    u1 = Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c3)
    u1 = concatenate([u1, c2])
    c4 = Conv2D(128, (3, 3), padding='same', kernel_initializer='he_normal')(u1)
    c4 = BatchNormalization()(c4)
    c4 = LeakyReLU(negative_slope=0.1)(c4)
    c4 = Conv2D(128, (3, 3), padding='same', kernel_initializer='he_normal')(c4)
    c4 = BatchNormalization()(c4)
    c4 = LeakyReLU(negative_slope=0.1)(c4)

    u2 = Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c4)
    u2 = concatenate([u2, c1])
    c5 = Conv2D(64, (3, 3), padding='same', kernel_initializer='he_normal')(u2)
    c5 = BatchNormalization()(c5)
    c5 = LeakyReLU(negative_slope=0.1)(c5)
    c5 = Conv2D(64, (3, 3), padding='same', kernel_initializer='he_normal')(c5)
    c5 = BatchNormalization()(c5)
    c5 = LeakyReLU(negative_slope=0.1)(c5)

    outputs = Conv2D(1, (1, 1), activation='sigmoid')(c5)

    model = Model(inputs, outputs)
    return model

# Reduce learning rate if validation loss does not improve
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=1, min_lr=1e-6)

# Create the U-Net model
unet = unet_model(input_shape=(224, 224, 3))
# unet.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['binary_accuracy'])
unet.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['binary_accuracy'])

# Train the model (Ensure X_train, y_train, X_val, y_val are defined)
history = unet.fit(
    x=X_train,
    y=y_train,
    validation_data=(X_val, y_val),
    batch_size=16,
    epochs=15,
    callbacks=[reduce_lr]
)

# Evaluate the model
loss, accuracy = unet.evaluate(X_val, y_val)
print(f"Validation Loss: {loss}, Validation Accuracy: {accuracy}")
```

```python
import tensorflow as tf
from tensorflow.keras.layers import (BatchNormalization, LeakyReLU, Dropout, Conv2D, MaxPooling2D,
                                     Conv2DTranspose, concatenate, Input)
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.optimizers import AdamW
from tensorflow.keras.metrics import MeanIoU
import tensorflow.keras.backend as K
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define Dice Loss
def dice_loss(y_true, y_pred):
    smooth = 1e-6
    intersection = K.sum(y_true * y_pred)
    return 1 - (2. * intersection + smooth) / (K.sum(y_true) + K.sum(y_pred) + smooth)

# U-Net Model
def unet_model(input_shape=(224, 224, 3)):
    inputs = Input(input_shape)

    # Encoder
    c1 = Conv2D(64, (3, 3), padding='same', kernel_initializer='he_normal')(inputs)
    c1 = BatchNormalization()(c1)
    c1 = LeakyReLU(negative_slope=0.1)(c1)
    c1 = Dropout(0.1)(c1)
    c1 = Conv2D(64, (3, 3), padding='same', kernel_initializer='he_normal')(c1)
    c1 = BatchNormalization()(c1)
    c1 = LeakyReLU(negative_slope=0.1)(c1)
    p1 = MaxPooling2D((2, 2))(c1)

    c2 = Conv2D(128, (3, 3), padding='same', kernel_initializer='he_normal')(p1)
    c2 = BatchNormalization()(c2)
    c2 = LeakyReLU(negative_slope=0.1)(c2)
    c2 = Dropout(0.2)(c2)
    c2 = Conv2D(128, (3, 3), padding='same', kernel_initializer='he_normal')(c2)
    c2 = BatchNormalization()(c2)
    c2 = LeakyReLU(negative_slope=0.1)(c2)
    p2 = MaxPooling2D((2, 2))(c2)

    c3 = Conv2D(256, (3, 3), padding='same', kernel_initializer='he_normal')(p2)
    c3 = BatchNormalization()(c3)
    c3 = LeakyReLU(negative_slope=0.1)(c3)
    c3 = Dropout(0.3)(c3)
    c3 = Conv2D(256, (3, 3), padding='same', kernel_initializer='he_normal')(c3)
    c3 = BatchNormalization()(c3)
    c3 = LeakyReLU(negative_slope=0.1)(c3)
    p3 = MaxPooling2D((2, 2))(c3)

    # Bottleneck
    c4 = Conv2D(512, (3, 3), padding='same', kernel_initializer='he_normal')(p3)
    c4 = BatchNormalization()(c4)
    c4 = LeakyReLU(negative_slope=0.1)(c4)
    c4 = Dropout(0.3)(c4)
    c4 = Conv2D(512, (3, 3), padding='same', kernel_initializer='he_normal')(c4)
    c4 = BatchNormalization()(c4)
    c4 = LeakyReLU(negative_slope=0.1)(c4)

    # Decoder
    u1 = Conv2DTranspose(256, (2, 2), strides=(2, 2), padding='same')(c4)
    u1 = concatenate([u1, c3])
    c5 = Conv2D(256, (3, 3), padding='same', kernel_initializer='he_normal')(u1)
    c5 = BatchNormalization()(c5)
    c5 = LeakyReLU(negative_slope=0.1)(c5)

    u2 = Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c5)
    u2 = concatenate([u2, c2])
    c6 = Conv2D(128, (3, 3), padding='same', kernel_initializer='he_normal')(u2)
    c6 = BatchNormalization()(c6)
    c6 = LeakyReLU(negative_slope=0.1)(c6)

    u3 = Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c6)
    u3 = concatenate([u3, c1])
    c7 = Conv2D(64, (3, 3), padding='same', kernel_initializer='he_normal')(u3)
    c7 = BatchNormalization()(c7)
    c7 = LeakyReLU(negative_slope=0.1)(c7)

    outputs = Conv2D(1, (1, 1), activation='sigmoid')(c7)

    model = Model(inputs, outputs)
    return model

# Compile Model
unet = unet_model(input_shape=(224, 224, 3))
unet.compile(optimizer=AdamW(learning_rate=0.0001),
             loss=lambda y_true, y_pred: 0.7 * dice_loss(y_true, y_pred) + 0.3 * tf.keras.losses.binary_crossentropy(y_true, y_pred),
             metrics=['binary_accuracy', MeanIoU(num_classes=2)])

# Callbacks
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=1, min_lr=1e-6)



# Train Model
history = unet.fit(
    X_train_resized, y_train_resized,
    validation_data=(X_val_resized, y_val_resized),
    batch_size=16,
    epochs=25,
    callbacks=[early_stop, reduce_lr]
)


# Evaluate Model
loss, accuracy, iou = unet.evaluate(X_val, y_val)
print(f"Validation Loss: {loss}, Validation Accuracy: {accuracy}, IoU: {iou}")
```

```python
import cv2
import numpy as np
import pandas as pd
from skimage.feature import graycomatrix, graycoprops
import os

def extract_features_from_masks(predicted_masks):
    features = []

    for mask_idx, mask in enumerate(predicted_masks):
        # Convert (H, W, 1) -> (H, W) if necessary
        if len(mask.shape) == 3 and mask.shape[2] == 1:
            mask = mask[:, :, 0]

        if len(mask.shape) == 3:
            mask = np.mean(mask, axis=-1)  # Convert 3-channel mask to single-channel
        h, w = mask.shape

        mask_binary = (mask > 0.5).astype(np.uint8)  # Binarize the mask

        # Find contours
        contours, _ = cv2.findContours(mask_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        for contour_idx, contour in enumerate(contours):
            area = cv2.contourArea(contour)
            perimeter = cv2.arcLength(contour, closed=True)
            x, y, w, h = cv2.boundingRect(contour)
            aspect_ratio = w / h if h != 0 else 0  # Avoid division by zero

            # Border Irregularity: Higher means more irregular shape
            if area > 0:
                border_irregularity = (perimeter ** 2) / (4 * np.pi * area)
            else:
                border_irregularity = 0

            # Compute symmetry score
            left_half = mask_binary[:, :w // 2]
            right_half = mask_binary[:, w // 2:]
            right_half_flipped = np.fliplr(right_half)

            # Ensure halves have the same shape
            min_h, min_w = min(left_half.shape[0], right_half_flipped.shape[0]), min(left_half.shape[1], right_half_flipped.shape[1])
            left_half, right_half_flipped = left_half[:min_h, :min_w], right_half_flipped[:min_h, :min_w]

            if np.sum(left_half | right_half_flipped) > 0:
                symmetry_score = np.sum(left_half & right_half_flipped) / np.sum(left_half | right_half_flipped)
            else:
                symmetry_score = 0

            # Extract Texture Features using Gray-Level Co-occurrence Matrix (GLCM)
            glcm = graycomatrix(mask_binary, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
            contrast = graycoprops(glcm, 'contrast')[0, 0]
            correlation = graycoprops(glcm, 'correlation')[0, 0]

            # Store features
            features.append({
                "mask_index": mask_idx,
                "contour_index": contour_idx,
                "area": area,
                "perimeter": perimeter,
                "aspect_ratio": aspect_ratio,
                "border_irregularity": border_irregularity,
                "symmetry_score": symmetry_score,
                "glcm_contrast": contrast,
                "glcm_correlation": correlation
            })

    return features

```

```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications.efficientnet import preprocess_input
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

# === Config ===
IMAGE_DIR = r"C:\Users\USTB-SE\Documents\Python Research\SymDerm_v2_DB\filtered_images"
EXCEL_PATH = r"C:\Users\USTB-SE\Documents\Python Research\SymDerm_v2_DB\SymDerm_v2_labels_with_all_malignancy.xlsx"
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS = 5

# === Load Excel File ===
df = pd.read_excel(EXCEL_PATH)

# Keep only necessary columns and valid classes
df = df[df['Malignancy_Text'].isin(['Benign', 'Malignant'])]
df = df[['File', 'Malignancy_Text']]

# Add full image path
df['File'] = df['File'].apply(lambda x: os.path.join(IMAGE_DIR, x))

# Train-Test Split
train_df, test_df = train_test_split(df, test_size=0.2, stratify=df['Malignancy_Text'], random_state=42)

# ImageDataGenerator with preprocessing
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input, horizontal_flip=True, zoom_range=0.2)
test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_gen = train_datagen.flow_from_dataframe(
    train_df,
    x_col='File',
    y_col='Malignancy_Text',
    target_size=IMAGE_SIZE,
    class_mode='binary',
    batch_size=BATCH_SIZE,
    shuffle=True
)

test_gen = test_datagen.flow_from_dataframe(
    test_df,
    x_col='File',
    y_col='Malignancy_Text',
    target_size=IMAGE_SIZE,
    class_mode='binary',
    batch_size=BATCH_SIZE,
    shuffle=False
)

# === Load Pre-trained Model ===
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(*IMAGE_SIZE, 3))
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
output = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=output)

# Freeze base model
base_model.trainable = False

model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# === Train ===
history = model.fit(
    train_gen,
    validation_data=test_gen,
    epochs=EPOCHS,
    callbacks=[EarlyStopping(patience=3, restore_best_weights=True)]
)

# === Evaluate ===
loss, acc = model.evaluate(test_gen)
print(f"Test Accuracy: {acc:.4f}")


# Predict using generator
y_pred_probs = model.predict(test_gen)
y_pred_classes = (y_pred_probs > 0.5).astype(int)

# Get true labels
y_true = test_gen.classes

# Classification Report
from sklearn.metrics import classification_report
print("\nClassification Report:")
print(classification_report(y_true, y_pred_classes, target_names=['Benign', 'Malignant']))

#confusion Matrix
cm = confusion_matrix(y_true, y_pred_classes)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

# Plot training & validation accuracy values
plt.figure(figsize=(12, 5))

# Accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='o')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# Loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss', marker='o')
plt.plot(history.history['val_loss'], label='Validation Loss', marker='o')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from torchvision import transforms
from PIL import Image

class AugmentedLesionDataset(Dataset):
    def __init__(self, images, labels):
        self.images = images
        self.labels = labels
        self.transform = transforms.Compose([
            transforms.ToTensor(),                      # [H, W, C] -> [C, H, W], scale [0,255] to [0,1]
            transforms.Resize((224, 224)),              # Resize for ViT
        ])

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img = self.images[idx]
        img = Image.fromarray(img.astype('uint8'))     # Convert numpy → PIL
        img = self.transform(img)
        label = self.labels[idx]
        return img, label


X_train, X_test, y_train, y_test = train_test_split(
    valid_images, valid_labels, test_size=0.2, random_state=42, stratify=valid_labels
)

train_dataset = AugmentedLesionDataset(X_train, y_train)
test_dataset = AugmentedLesionDataset(X_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32)
```
